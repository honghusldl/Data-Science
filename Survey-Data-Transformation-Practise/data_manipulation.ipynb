```python
import pandas as pd
import os
import numpy as np
```

```python
pwd = os.getcwd()
```

```python
path = pwd + "/Data - Survey Monkey Output.xlsx"
df = pd.read_excel(path, sheet_name="edited_data")
df
```

```python
df_modified = df.copy()
```

```python
df_modified.columns
```

```python
columns_to_drop = ['Start Date_', 'End Date_', 'Email Address_','First Name_', 'Last Name_', 'Custom Data 1_']
```

```python
df_modified.drop(columns=columns_to_drop,inplace=True)
df_modified
```

```python
# unpivoting question using pd.melt
id_vars = list(df_modified)[:8] # first 8 column as idenfitiers
value_vars = list(df_modified)[8:] # unpivoting the rest columns
```

```python
df_melted = df_modified.melt(id_vars=id_vars,value_vars=value_vars,value_name="Response",var_name="Questions_Subquestions")
df_melted
```

```python
df_question = pd.read_excel(path, sheet_name="questions")
df_question
```

```python
questions = df_question.copy()
questions = questions[['question','question_subquestion']]
questions.columns = ['Questions','Questions_Subquestions']
questions.dropna(inplace=True)
questions
```

```python
df_merged = pd.merge(df_melted,questions,how = "left",on = "Questions_Subquestions");
print(len(df_merged) == len(df_melted))
df_merged
```

```python
# first we exclude all Null values in Column Response
actual_responses = df_merged[df_merged["Response"].notna()]

# count number of respondents for each question
actual_responses = actual_responses.groupby("Questions")["Respondent ID_"].nunique().reset_index()

actual_responses.rename(columns = {'Respondent ID_' : 'Respondent_Count'},inplace=True)
actual_responses
```

```python
df_merged_num_of_respondent = pd.merge(df_merged,actual_responses,how = "left",on = "Questions");
print(len(df_merged) == len(df_melted))
df_merged_num_of_respondent
```

```python
# we need to include null values for each question as a category (no response)
same_answer = df_merged # [df_merged["Response"].notna()]

# count number of respondents for each question
same_answer = same_answer.groupby(["Questions_Subquestions","Response"])["Respondent ID_"].nunique().reset_index()

same_answer.rename(columns = {'Respondent ID_' : 'Respondent_Count'},inplace=True)
same_answer
```

```python
df_merged_same_answer = pd.merge(df_merged_num_of_respondent,same_answer,how = "left",on = ["Questions_Subquestions","Response"]);
print(len(df_merged_num_of_respondent) == len(df_merged_same_answer))
df_merged_same_answer
```

```python
output = df_merged_same_answer.copy()
output.rename(columns = 
    {"Identify which division you work in._Response":"Division Primary",
    "Identify which division you work in._Other (please specify)":"Division Secondary",
    "Which of the following best describes your position level?_Response":"Position",
    "Which generation are you apart of?_Response":"Generation",
    "Please select the gender in which you identify._Response":"Gender",
    "Which duration range best aligns with your tenure at your company?_Response":"Tenure",
    "Which of the following best describes your employment type?_Response":"Employment Type"},inplace=True)
output
```

```python
output.to_excel(pwd + "/ouput.xlsx", index=False)
```
